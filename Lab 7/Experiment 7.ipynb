### Problem 2.1 -Preparing the Corpus

```python
corpus = df.copy()
```

```python
corpus['text'] = corpus['text'].str.replace('[^a-zA-Z#]', " ")
corpus
```

```python
corpus['text']
corpus_arr = list(map(lambda x: x.lower(), corpus['text']))
len(corpus_arr)
```

```python
corpus_without_sw = []
for sentence in corpus_text:
  corpus_without_sw.append(' '.join([word for word in sentence.split() if not word in stop_words]))
corpus_without_sw
```

```python
corpus = corpus_without_sw
len(corpus)
```

##  Stemmming of Words

```python
porter = PorterStemmer()
corpus_stemmed = []
for sentence in corpus:
  corpus_stemmed.append(' '.join([porter.stem(word) for word in sentence.split()]))

corpus_stemmed
```

```python
corpus = corpus_stemmed
len(corpus)
```

```python
df['texts'] = corpus
```

# Building Document Term Matrix

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['texts'], df['spam'], test_size=0.33, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfvec = TfidfVectorizer()
X_train = tfvec.fit_transform(X_train)
X_test = tfvec.transform(X_test)
```
